# Carga de datos con open

```python
data3 = open(mainpath+"/"+"customer-churn-model/Customer Churn Model.txt","r")
```

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
cols
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

# Lectura y escritura de ficheros

```python
infile = mainpath+"/"+"customer-churn-model/Customer Churn Model.txt"
outfile = mainpath+"/"+"customer-churn-model/Customer Churn Model Tab.txt"
```

```python
with open(infile,"r") as infile1:
    with open(outfile,"w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

# Lectura de datos por medio de url

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
medals_data.head()
```

# Tarea
Realizar la lectura de de los datos que nos viene en la url del csv anterior por medio de la libreria de urllib3.

```python
import urllib3
http = urllib3.PoolManager()
r = http.request('GET',medals_url)
r.status
response = r.data.decode("utf-8")
```

```python
count = 0
url_data = response.split("\n");
url_cols = url_data[0].split(",")
url_dict = {}
for col in url_cols:
    url_dict[col] = []
for line in url_data:
    if count > 0:
        values = line.split(",")
        for i in range(len(url_cols)):
            url_dict[url_cols[i]].append(values[i]) 
    count += 1
```

```python
dft = pd.DataFrame(url_dict)
dft.to_csv(mainpath+"/"+"/titanic/medals.csv")
dft.to_excel(mainpath+"/"+"/titanic/medals.xls")
dft.to_json(mainpath+"/"+"/titanic/medals.json")
dft
```
